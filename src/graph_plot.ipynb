```python
import matplotlib.pyplot as plt
```

```python
x=[i for i in range(1,11)]
c=[]
a=[]
b=[]
for i in x:
    count1=0
    count2=0
    for j in range(0,2**i):
        for k in range(0,8):
            if j==k:
                count1+=1
            if j<k:
                count2+=1
    collision=count1/((2**i)*8)
    awin=count2/((2**i)*8)
    bwin=1-awin-collision
    c.append(collision)
    a.append(awin)
    b.append(bwin)
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(x, c)
ax.set_ylabel('Collision')
ax.set_xlabel('Collision number (n)')
```

```python
c1=[]
a1=[]
b1=[]
for i in x:
    count1=0
    count2=0
    for j in range(0,2**i):
        for k in range(0,8):
            if (j/i)==(k/3):
                count1+=1
            if (j/i)<(k/3):
                count2+=1
    collision=count1/((2**i)*8)
    awin=count2/((2**i)*8)
    bwin=1-awin-collision
    c1.append(collision)
    a1.append(awin)
    b1.append(bwin)
```

```python
fig, ax = plt.subplots(figsize=(16, 9))
ax.plot(x, c)
ax.plot(x,c1)
ax.legend(["BEB","SBEB"])
ax.set_ylabel('Collision')
ax.set_xlabel('Collision number (n)')
```
